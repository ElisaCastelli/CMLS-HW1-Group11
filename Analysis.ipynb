# Analysis

```python
import numpy as np
import librosa
import scipy as sp
%matplotlib inline

import matplotlib.pyplot as plt
import IPython.display as ipd
import scipy.stats
import seaborn as sns
import random
import csv
```

```python
class Sound:
    
    def __init__(self, path, slice_file_name, fold, classID, class_name, startTime, endTime):
        self.slice_file_name = path +'/fold'+fold+'/'+slice_file_name
        self.fold = fold
        self.classID = classID
        self.class_name = class_name
        self.startTime = startTime
        self.endTime = endTime
        
    def __getDuration__(self):
        return (float(self.endTime) - float(self.startTime))
```

```python
## read csv

classes = ['air_conditioner',
           'car_horn',
           'children_playing',
           'dog_bark',
           'drilling',
           'engine_idling',
           'gun_shot',
           'jackhammer',
           'siren',
           'street_music'
          ]

path = 'UrbanSound/UrbanSound8K.csv'
soundList = []
with open(path, newline='') as soundcsv:
            reader = csv.DictReader(soundcsv)
            for row in reader:
                s = Sound('UrbanSound/', row['slice_file_name'], row['fold'], row['classID'], row['class'], row['start'], 
                          row['end'])
                #if s.__getDuration__() == 4.0:
                soundList.append(s)
```

```python
fileForClass = []
for i in range(1, len(classes)+1):
    files = [s for s in soundList if s.class_name == classes[i-1]] 
    fileForClass.append(len(files))

print(fileForClass)
```

```python
for i in range(1,1000):
    j = random.randint(1, len(soundList))
    example, fs = librosa.load(soundList[j].slice_file_name, sr=None)

    print(soundList[j].slice_file_name + " - duration: " + str(example.shape[0]))
```

```python
plt.figure(figsize=(16, 16))

for i in range(1, len(classes)+1):
    plt.subplot(5,2,i)
    classSound = [s for s in soundList if s.class_name == classes[i-1]] 
    j = random.randint(1, len(classSound))
    audioExample, fs = librosa.load(classSound[j].slice_file_name, sr=None)
    time_axis = np.arange(audioExample.shape[0]) / fs
    plt.plot(time_axis, audioExample)
    plt.grid(True)
    plt.title(classes[i-1])
    
    
```

```python
plt.figure(figsize=(16, 16))

for i in range(1, len(classes)+1):
    plt.subplot(5,2,i)
    classSound = [s for s in soundList if s.class_name == classes[i-1]] 
    j = random.randint(1, len(classSound))
    audioExample, fs = librosa.load(classSound[j].slice_file_name, sr=None)
    time_axis = np.arange(audioExample.shape[0]) / fs
    plt.magnitude_spectrum(audioExample,fs)
    plt.grid(True)
    plt.title(classes[i-1])
```

```python
#Zero Crossing Rate
plt.figure(figsize=(16, 16))

for i in range(1, len(classes)+1):
    plt.subplot(5,2,i)
    classSound = [s for s in soundList if s.class_name == classes[i-1]] 
    j = random.randint(1, len(classSound))
    audioExample, fs = librosa.load(classSound[j].slice_file_name, sr=None)
    ZCR = librosa.feature.zero_crossing_rate(y=audioExample)
    time_axis = np.arange(ZCR.shape[1]) *512/fs
    plt.plot(time_axis,ZCR[0])
    plt.grid(True)
    plt.title(classes[i-1])
```

```python
#Spectral Centroid
plt.figure(figsize=(16, 16))

for i in range(1, len(classes)+1):
    plt.subplot(5,2,i)
    classSound = [s for s in soundList if s.class_name == classes[i-1]] 
    j = random.randint(1, len(classSound))
    audioExample, fs = librosa.load(classSound[j].slice_file_name, sr=None)
    SC = librosa.feature.spectral_centroid(y=audioExample, sr=fs)
    time_axis = np.arange(SC.shape[1]) *512/fs
    plt.plot(time_axis, SC[0])
    plt.grid(True)
    plt.title(classes[i-1])

       
```
